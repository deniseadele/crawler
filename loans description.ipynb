```python
# scrape loan description for first 250,000 loans

# Using Kiva API
df = pd.read_csv('all_loans_1.csv', header = 0)
n = 0
session = requests.Session()

df['description'] = ''

for i, row in df.iterrows():
    try:

        loanid = row['LOAN_ID']
        
        # Get HTML
        url = 'https://www.kiva.org/lend/' + str(loanid) + '?minimal=false'
        # https://www.kiva.org/lend/1114506?minimal=false
        
        r = session.get(url)
        soup = BeautifulSoup(r.text, 'html.parser')

        description = soup.find('section', class_= 'loan-description')
        description2 = " ".join(description.text.replace('\r',' ').replace('\n',' ').strip().split())

        df.at[i,'description'] = description2
        
    
    # error handling
    except AttributeError:
        df.at[i,'description'] = ''
        continue
    except ValueError:
        df.at[i,'description'] = ''
        continue
    except requests.exceptions.Timeout:
        df.at[i,'description'] = 'timeout'
    except requests.exceptions.TooManyRedirects:
        df.at[i,'description'] = 'too many redirects'
    except requests.exceptions.RequestException:
        df.at[i,'description'] = 'request exception'
    except requests.exceptions.HTTPError:
        df.at[i,'description'] = 'HTTP Error'
    except requests.exceptions.ConnectionError:
        df.at[i,'description'] = 'Connection Error'
    except requests.exceptions.Timeout:
        df.at[i,'description'] = 'Timeout Error'
    except:
        df.at[i,'description'] = 'Some error'


df.to_csv('all_loans_1_parse.csv', header=True, index=None)
```
